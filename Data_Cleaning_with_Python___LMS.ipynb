### Extracting Dataset

```python
# import libraries

import pandas as pd
```

```python
# create a dataset

url_link1 = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
company_data = pd.read_html(url_link1)
```

```python
type(company_data)
```

```python
company_data = company_data[0]
```

```python
type(company_data)
```

```python
company_data.head()
```

```python
company_data.columns
```

```python
company_data.columns = company_data.columns.get_level_values(0)
```

```python
company_data.columns
```

```python
company_data.head()
```

```python
# dropping the ref column

company_data.drop(columns=['Ref.', 'State-owned'], inplace=True)
```

### Renaming the columns

```python
company_data.columns
```

```python
company_data.rename(columns =
                    {'Rank': 'Ranking',
                     'Name': 'Company Name',
                     'Revenue': 'Revenue(USD millions)',
                     'Profit': 'Profit(USD millions)',
                     'Headquarters[note 1]': 'Headquarters',
                    'Unnamed: 9_level_0': 'Revenue per worker'}, inplace=True
                    )
```

```python
company_data.head()
```

```python
# remove the dollar sign and comma from revenue and profit

company_data['Revenue(USD millions)'] = company_data['Revenue(USD millions)'].str.replace('[\$,]', '', regex=True).astype(float)
company_data['Profit(USD millions)'] = company_data['Profit(USD millions)'].str.replace('[^\d−]', '', regex=True)
```

```python
# change the minus sign to a standard minus sign
# change the 'Profit(USD millions) to numeric values while omitting the empty column

company_data['Profit(USD millions)'] = company_data['Profit(USD millions)'].str.replace('−', '-', regex=True)
company_data['Profit(USD millions)'] = pd.to_numeric(company_data['Profit(USD millions)'], errors='coerce')
```
