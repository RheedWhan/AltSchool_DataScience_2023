```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv('/content/drive/MyDrive/my csv/train.csv')
titanic_data.head()
```

```python
titanic_data.shape
```

```python
titanic_data.info()
```

```python
# 1. What is the overall survival rate on the Titanic
titanic_data['Survived'].mean()
```

```python
# 2. How does survival rate differs between gender
titanic_data.groupby('Sex')['Survived'].mean()
```

```python
# 3a. What is average age of survivals
titanic_data[titanic_data['Survived'] == 1]['Age'].mean()
```

```python
# 3b. What is average age of non-survivals
titanic_data[titanic_data['Survived'] == 0]['Age'].mean()
```

```python
# Summary of the data
titanic_data.describe()
```

```python
# 4. Are there any significant difference in the survival based on class
titanic_data.groupby('Pclass')['Survived'].mean()
```

```python
# 5. How does the survival rate vary from different age group (e.g childrens, adults, seniors)
age_groups = pd.cut(titanic_data['Age'], bins=[0, 18, 35, 60, float('inf')],
                    labels = ['child', 'Adults', 'Middle Aged', 'Senior'])
titanic_data.groupby(age_groups)['Survived'].mean()
```
