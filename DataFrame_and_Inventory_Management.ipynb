# Inventory Management

Build an inventory management system for a supermarket

```python
import pandas as pd
```

```python
# add product
def add_product():
    product_name = input('Enter the name of the product: ')
    quantity = int(input('Enter the quantity: '))
    price = float(input('Enter the price: '))

    with open('inventory_data.csv', 'a') as file:
        file.write(f'{product_name}, {quantity}, {price} \n')
    print(f'{product_name} added successfully!\n')


# view inventory
def view_inventory():
    try:
        df = pd.read_csv('inventory_data.csv', header=None, names = ['Product Name', 'Quantity', 'Price'])

        if df.empty:
            print('No product in the inventory')
        else:
            print('\n')
            print('Product Inventory: ')
            display(df)
    except FileNotFoundError:
        print('File does not exist')



# remove a product
def remove_product():
    remove_name = input('Enter a name of the product to remove: ')

    try:
        df = pd.read_csv('inventory_data.csv', header=None, names=['Product Name', 'Quantity', 'Price'])

        remove = df[df['Product Name'].str.lower() == remove_name.lower()].drop(remove_name, axis=0)

        if remove.empty:
            print(f"No product with the name: {remove_name}")
        else:
            print('product found')
            return remove
    except FileNotFoundError:
        print('file not found')




# search for product
def search_product():
    search_name = input('Enter a name of the product to search: ')

    try:
        df = pd.read_csv('inventory_data.csv', header=None, names=['Product Name', 'Quantity', 'Price'])

        result = df[df['Product Name'].str.lower() == search_name.lower()]

        if result.empty:
            print(f"No product with the name: {search_name}")
        else:
            print('product found')
            return result
    except FileNotFoundError:
        print('file not found')


# run inventory
def run_inventory_ms():
    while True:
        print('\nInventory Management System')
        print('1. Add Product')
        print('2. Remove Product')
        print('3. View Inventory')
        print('4. Search Inventory')
        print('5. Exit')

        choice = input('Enter your choice (1-4): ')
        if choice == '1':
            add_product()
        elif choice == '2':
            remove_product()
        elif choice == '3':
            view_inventory()
        elif choice == '4':
            search_product()
        elif choice == '4':
            break
        else:
            print('Invalid choice. please enter a number between 1 and 4.')
```

```python
run_inventory_ms()
```
