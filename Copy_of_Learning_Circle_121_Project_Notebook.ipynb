#### Economic Growth

- How has the GDP at constant market prices changed over the years?

```python
df.groupby('Year')['GDP at 2010 Constant Market Prices'].mean()
```

**Takeaway:** The GDP at constant arket prices has been increasing over time.
There was little setback in the year 2020 probably because of COVID-19, but it was back on track the following year

- Are there any specific periods of economic growth or recession?

```python
# Identify economic growth and recession periods based on GDP trend
# Using 2% as a differentiator between the current GDP and the previous one
threshold_growth = 0.02  # 2% growth
threshold_recession = -0.02  # 2% contraction

# Create a new column indicating economic status (growth, recession, or stable)
df['Economic_Status'] = 'Stable'
df.loc[df['GDP at 2010 Constant Market Prices'].pct_change() > threshold_growth, 'Economic_Status'] = 'Growth'
df.loc[df['GDP at 2010 Constant Market Prices'].pct_change() < threshold_recession, 'Economic_Status'] = 'Recession'

# Print periods of economic growth and recession
growth_periods = df[df['Economic_Status'] == 'Growth']['Year'].tolist()
recession_periods = df[df['Economic_Status'] == 'Recession']['Year'].tolist()
```

```python
df.head() # Check the new column 'Economic_Status'
```

```python
# Print years with growth or recession
print("Periods of Economic Growth:", growth_periods)
print("Periods of Economic Recession:", recession_periods)
```
